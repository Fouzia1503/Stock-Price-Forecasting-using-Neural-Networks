```python
# Since stock prediction is a 'Univariate time series data', retrieve the 'Adj Close' column alone from the dataset and converting it into a numoy array and reshaping the array
mod_data = np.array(stock_data['Adj Close']).reshape(-1,1)
mod_data.shape
```

```python
#Train-Test Split
#Splitting the 'Adj Close' column values into train and test datasets
# Allocating 70% of data to the train and remaining to the test
training_data_num = len(mod_data) * 0.7
training_data_num
```

```python
import math
training_data_num=math.ceil(training_data_num)
print(training_data_num)
```

```python
training_data = mod_data[0:training_data_num]
training_data.shape
```

```python
training_data
```

```python
# Scaling the train data using the minmaxscaler so that the values lies between the range 0 and 1.
scaled_values = MinMaxScaler(feature_range=(0,1))
train_scale=scaled_values.fit_transform(training_data)
```

```python
train_scale.shape
```

```python
# Creating a 45 window to predict the next day's stock price
# In time series data, the independent values depend on the values that are preceeding
# Dependent values are the feature data and independent values are the label data
# X_train stores the feature data and y_train stores the label data
# Convert x_train and y_train into numpy arrays and reshape them
# Same applies for test split
X_train =[]
y_train =[]

for n in range(45,len(train_scale)-1):
    X_train.append(train_scale[n-45:n,0])
    y_train.append(train_scale[n,0])
    if n <=46:
        print(X_train)
        print(y_train)

    
```

```python
X_train = np.array(X_train)
X_train.shape
```

```python
y_train=np.array(y_train)
y_train.shape
```

```python
# The LSTM model accepts only three dimension
# Modify the shape of the array to 3 Dimension
X_train = np.reshape(X_train, (X_train.shape[0],X_train.shape[1],1))
```

```python
y_train = np.reshape(y_train, (y_train.shape[0],1,1))
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
# Allocating the remaining values of the 'Adj Close' column as testing dataset
testing_data = mod_data[training_data_num-45:,:]
testing_data.shape
```

```python
testing_data
```

```python
test_scale=scaled_values.fit_transform(testing_data)
```

```python
test_scale.shape
```

```python

X_test =[]
y_test = mod_data[training_data_num:,:]

for n in range(45,len(test_scale)):
    X_test.append(test_scale[n-45:n,0])
    if n <=46:
        print(X_test)
        
```

```python
X_test = np.array(X_test)
X_test.shape
```

```python
X_test = np.reshape(X_test, (X_test.shape[0],X_test.shape[1],1))
X_test.shape
```

```python
#Building the LSTM model for the training data
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
```

```python
lstm = Sequential()
lstm.add(LSTM(30,return_sequences=True,input_shape=(X_train.shape[1],1)))
lstm.add(LSTM(30,return_sequences=False))
lstm.add(Dense(15))
lstm.add(Dense(1))
```

```python
#Parameter tuning for optimization of the model for better performance
lstm.compile(optimizer='adam', loss='mean_squared_error')
```

```python
lstm.fit(X_train,y_train,batch_size=1,epochs=6)
```

```python
# Evaluating the model by predicting the testing data
predictions = lstm.predict(X_test)
# Scaler inverse for unscaling the data to original format 
predictions = scaled_values.inverse_transform(predictions)
```

```python
# Root mean squared error is a statsitic metric  
# Find out the variablity of values between the actual datapoints and predicted datapoints
# Estimating the root mean squared error for testing data
# Computing RMSE for six repetitions and finding the mean
mean_error = []
for i in range(3):
    lstm.fit(X_train,y_train,batch_size=1,epochs=6)
    predictions = lstm.predict(X_test)
    predictions = scaled_values.inverse_transform(predictions)
    rmse = np.sqrt( np.mean(predictions - y_test)**2)
    mean_error.append(rmse)
# Lower values of RMSE indicates that the model has performed well!
```

```python
np.array(mean_error).reshape(3,1)
np.mean(mean_error)
```
